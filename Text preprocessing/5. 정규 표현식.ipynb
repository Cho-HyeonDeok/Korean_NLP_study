# 정규 표현식 실습

```python
import re
```

```python
# 1) 기호
# a c 사이엔 아무 문자나 올 수 있음
r = re.compile("a.c")
r.search("kkk") # 아무런 결과도 출력되지 않는다.
```

```python
r.search("abc")
```

```python
# 2) ?기호
# b가 있ㅇ르수도 없을 수도 있음
r = re.compile("ab?c")
r.search("abbc") # 아무런 결과도 출력되지 않는다.
```

```python
r.search("abc")
```

```python
# 3) *기호
# * 바로 앞 문자가 0개 이상일때
r = re.compile("ab*c")
r.search("a") # 아무런 결과도 출력되지 않는다.
```

```python
r.search("ac")
```

```python
r.search("abc") 
```

```python
r.search("abbbbc") 
```

```python
# + 기호
# + 앞 문자가 최소 1개 이상
r = re.compile("ab+c")
r.search("ac") # 아무런 결과도 출력되지 않는다.
```

```python
r.search("abc") 
```

```python
r.search("abbbbc") 
```

```python
# 5) ^기호
# 시작되는 문자열 지정 ^ab면 ab로 시작하는 단어만
r = re.compile("^ab")

# 아무런 결과도 출력되지 않는다.
r.search("bbc")
r.search("zab")
```

```python
r.search("abz")
```

```python
# 6) {숫자} 기호
# 해당 숫자만큼 반복한 것
r = re.compile("ab{2}c")

# 아무런 결과도 출력되지 않는다.
r.search("ac")
r.search("abc")
r.search("abbbbbc")
```

```python
r.search("abbc")
```

```python
# 7) {숫자1, 숫자2} 기호
# 숫자 1 이상, 2이하 반복
r = re.compile("ab{2,8}c")

# 아무런 결과도 출력되지 않는다.
r.search("ac")
r.search("abc")
r.search("abbbbbbbbbc")
```

```python
r.search("abbc")
```

```python
r.search("abbbbbbbbc")
```

```python
# 8) {숫자,} 기호
# 해당 문자를 숫자 이상 만큼 반복
# a{2,}bc 면 뒤에 bc가 붙으면서 a가 2개 이상인 것
r = re.compile("a{2,}bc")

# 아무런 결과도 출력되지 않는다.
r.search("bc")
r.search("aa")
```

```python
r.search("aabc")
```

```python
r.search("aaaaaaaabc")
```

```python
# 9) [ ] 기호
# [] 안의 문자가 있으면 매치
r = re.compile("[abc]") # [abc]는 [a-c]와 같다.
r.search("zzz") # 아무런 결과도 출력되지 않는다.
```

```python
r.search("a")
```

```python
r.search("aaaaaaa")                  
```

```python
r.search("baac") 
```

```python
# 소문자 전체 예시 
r = re.compile("[a-z]")

# 아무런 결과도 출력되지 않는다.
r.search("AAA")
r.search("111") 
```

```python
r.search("aBC")
```

```python
# 10) [^문자] 기호
# ^ 뒤에 문자 제외 나머지 모든 문자 
r = re.compile("[^abc]")

# 아무런 결과도 출력되지 않는다.
r.search("a")
r.search("ab") 
r.search("b")
```

```python
r.search("d")
```

```python
r.search("1")      
```

# 3. 정규 표현식 모듈 함수 예제

```python
# (1) re.match() 와 re.search()의 차이
# match는 시작부터 같아야함

r = re.compile("ab.")
r.match("kkkabc") # 아무런 결과도 출력되지 않는다.
```

```python
r.search("kkkabc")  
```

```python
r.match("abckkk")  
```

```python
# (2) re.split()
# 분리할때 사용 토큰화 주요 함수
# 공백 기준 분리
text = "사과 딸기 수박 메론 바나나"
re.split(" ", text)
```

```python
# 줄바꿈 기준 분리
text = """사과
딸기
수박
메론
바나나"""

re.split("\n", text)
```

```python
# '+'를 기준으로 분리
text = "사과+딸기+수박+메론+바나나"

re.split("\+", text)
```

```python
# (3) re.findall()
# 정규 표현식 매치되는 모든 문자 리스트로 리턴
text = """이름 : 김철수
전화번호 : 010 - 1234 - 1234
나이 : 30
성별 : 남"""

re.findall("\d+", text)
```

```python
re.findall("\d+", "문자열입니다.")
```

```python
# (4) re.sub()
# 일치하는 문자열을 다른 문자열으로 대체 ex) 알파벳 외 문자는 공백으로 대체
text = "Regular expression : A regular expression, regex or regexp[1] (sometimes called a rational expression)[2][3] is, in theoretical computer science and formal language theory, a sequence of characters that define a search pattern."

preprocessed_text = re.sub('[^a-zA-Z]', ' ', text)
print(preprocessed_text)
```

# 4. 정규 표현식 텍스트 전처리 예제


```python
text = """100 John    PROF
101 James   STUD
102 Mac   STUD"""
```

```python
# \s+ 는 뒤에붙는 공백을 최소 1개이상 찾아내는 정규 표현식
re.split('\s+', text)  
```

```python
# \d+ 는 숫자만 뽑아오기 d는 숫자 +는 한개 이상의 숫자
re.findall('\d+',text)  
```

```python
# 대문자만
re.findall('[A-Z]',text)
```

```python
# 대문자가 4번 나올때만
re.findall('[A-Z]{4}',text)  
```

```python
# 대문자 + 소문자 1개이상
re.findall('[A-Z][a-z]+',text)
```

# 5. 정규 표현식을 이용한 토큰화


```python
from nltk.tokenize import RegexpTokenizer

text = "Don't be fooled by the dark sounding name, Mr. Jone's Orphanage is as cheery as cheery goes for a pastry shop"

tokenizer1 = RegexpTokenizer("[\w]+") # 문자 또는 숫자가 1개 이상일때
tokenizer2 = RegexpTokenizer("\s+", gaps=True) # 띄어쓰기 1개 이상일때 / gaps=true는 공백이 토큰화 기준이라는 것을 나타냄 안쓰면 공백만 추출

print(tokenizer1.tokenize(text))
print(tokenizer2.tokenize(text))
```
